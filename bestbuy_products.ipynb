```python
import requests
import json
import pandas as pd
```

```python
api_key = 'key'
product_name = "Apple 8 GPS Cell 45mm Silver Aluminum Sports Band M L"
url = f"https://api.bestbuy.com/v1/products((search={product_name}))?format=json&apiKey={api_key}"
```

```python
response = requests.get(url)
```

```python
data = json.loads(response.text)
df = pd.DataFrame(data)
df
```

```python
df.to_csv('product_data.csv', index=False)
```

```python
product_info = df["products"][0]
product_info
```

```python
products = data['products']
for product in products:
    priceUpdateDate = product['priceUpdateDate']
    itemUpdateDate = product['itemUpdateDate']
    active = product['active']
    name = product['name']
    price = product['salePrice']
    discount = product['percentSavings']
    availability = product['onlineAvailability']
    sku = product['sku']

    # Print the extracted information
    print(f"priceUpdateDate: {priceUpdateDate}")
    print(f"itemUpdateDate: {itemUpdateDate}")
    print(f"Active: {active}")
    print(f"Product name: {name}")
    print(f"Price: {price}")
    print(f"Discount: {discount}")
    print(f"Availability: {availability}")
    print(f"SKU: {sku}")
    print("-----------------------")
```

```python
selected_items = ["priceUpdateDate","sku", "name", "salePrice", "percentSavings", "onlineAvailability","url", "addToCartUrl"]
```

```python
data_dict = {item: [x[item] for x in products] for item in selected_items}
products_df = pd.DataFrame(data_dict)
products_df
```

```python
products_df.to_csv('bestbuy.csv', index=False)
```
